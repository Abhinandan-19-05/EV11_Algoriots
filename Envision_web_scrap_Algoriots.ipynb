```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
```

## Fetch Dataset of Liver Disorder

The **Liver Disorders dataset** is available from the UCI Machine Learning Repository.  
This dataset contains blood test results of individuals and is commonly used in regression and exploratory analysis tasks. :contentReference[oaicite:0]{index=0} </br></br>


## ðŸ“¥ Download from UCI Repository

You can manually download the dataset files from:

ðŸ”— https://archive.ics.uci.edu/dataset/60/liver+disorders

```python
# fetch dataset
liver_disorders = fetch_ucirepo(id=60)

# data (as pandas dataframes)
X = liver_disorders.data.features
y = liver_disorders.data.targets
```

```python
df_liver_disorder = pd.concat([X, y], axis=1)
df_liver_disorder.head(5)
```
